```python
import numpy as np 
import pandas as pd
import os
```

```python
# download from ChEMBL

with open('DOWNLOAD.csv', 'r') as f:
    data_list = f.read().strip().split('\n')
```

```python
Chembl_IDs, Smiles = [], []
for i in range(1, len(data_list)):
    data2 = data_list[i].split('"')
    Chembl_ID = data2[1].strip()
    Smile = data2[-2].strip()
    Chembl_IDs.append(Chembl_ID)
    Smiles.append(Smile)
```

```python
df = pd.DataFrame(Smiles, columns=['drug'])
df.head()
```

```python
df2 = pd.DataFrame(Chembl_IDs, columns=['Chembl_ID'])
df2.head()
```

```python
df3 = pd.concat([df2, df], axis=1)
df3.head()
```

```python
df3.info()
```

```python
df3.replace(to_replace=r'^\s*$',value=np.nan,regex=True,inplace=True)
df3
```

```python
df4=df3.dropna(axis=0,how='any')
```
